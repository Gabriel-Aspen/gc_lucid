Testing out violin and tsnetools on various datasets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.manifold import TSNE

from violin import numeric, categorical
from tsnetools import categorical_tsne, numeric_tsne
```

```python
#House votes dataset
# columns: Particular bills (immigration, education spending, etc.) - described as Y/N
# target: republican (1) or not (0)
dataset_url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/voting-records/house-votes-84.data'
data = pd.read_csv(dataset_url, sep=',')
data.columns = data.columns.str.replace(' ', '_')
```

```python
categorical(data, target = 'republican')
```

```python
categorical_tsne(data, target = 'republican')
```

```python
#wine dataset
# columns: particular attributes of wine (density, alcohol, etc.)
# target: quality rating given by wine experts
dataset_url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
data = pd.read_csv(dataset_url, sep=';')
data.columns = data.columns.str.replace(' ', '_')
```

```python
numeric(data, target = 'quality')
```

```python
numeric_tsne(data, target = 'quality')
```

```python
#mushrooms
# columns: Logical rules for mushroom identification (cap shape, spore print color, etc.)
# target: poisonous (1) or edible (0)
dataset_url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data'
data = pd.read_csv(dataset_url, sep=',')
data.columns = data.columns.str.replace(' ', '_')
```

```python
categorical(data, target = 'p')
```

```python
categorical_tsne(data, target = 'p')
```

```python
#yeast protein localization sites
# columns: ** Complicated, see below**
# target: Location site of protein (nucleus, mitochondria, etc.)
dataset_url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/yeast/yeast.data'
data = pd.read_csv(dataset_url, sep= "\s+", 
                   names=['Sequence_name', 'mcg', 'gvh', 'alm', 'mit', 'erl', 'pox',
                          'vac', 'nuc', 'Site'])
```

```python
categorical(data, target='Site')
```

```python
categorical_tsne(data, target='Site')
```

Description of the columns for yeast
  1.  Sequence Name: Accession number for the SWISS-PROT database
  2.  mcg: McGeoch's method for signal sequence recognition.
  3.  gvh: von Heijne's method for signal sequence recognition.
  4.  alm: Score of the ALOM membrane spanning region prediction program.
  5.  mit: Score of discriminant analysis of the amino acid content of
	   the N-terminal region (20 residues long) of mitochondrial and 
           non-mitochondrial proteins.
  6.  erl: Presence of "HDEL" substring (thought to act as a signal for
	   retention in the endoplasmic reticulum lumen). Binary attribute.
  7.  pox: Peroxisomal targeting signal in the C-terminus.
  8.  vac: Score of discriminant analysis of the amino acid content of
           vacuolar and extracellular proteins.
  9.  nuc: Score of discriminant analysis of nuclear localization signals
	   of nuclear and non-nuclear proteins.
